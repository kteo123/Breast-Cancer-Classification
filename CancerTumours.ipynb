```python
import pandas as pd
```

```python
cancer = pd.read_csv('/content/Cancer_Data.csv')
```

```python
cancer_cleaned = cancer.drop(columns=['id', 'radius_mean', 'perimeter_mean', 'area_mean', 'radius_se', 'perimeter_se', 'area_se', 'radius_worst', 'perimeter_worst', 'area_worst', 'Unnamed: 32'])
```

```python
cancer_cleaned.columns
```

```python
cancer_with_diagnosis = pd.get_dummies(cancer_cleaned, columns=['diagnosis',])
cancer_with_diagnosis.dropna()
```

```python
cancer_with_diagnosis['diagnosis_M'] = cancer_with_diagnosis['diagnosis_M'].astype(int)
cancer_encoded = cancer_with_diagnosis.drop(columns=['diagnosis_B'])
cancer_encoded
```

```python
import matplotlib.pyplot as plt
```

```python
correlation_matrix = cancer_encoded.corr()

plt.figure(figsize=(10, 8))
plt.imshow(correlation_matrix, cmap='magma', interpolation='nearest')
plt.colorbar(label='Correlation Coefficient')
plt.title('Correlation Matrix of season_finalists DataFrame')
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.tight_layout()
plt.show()
```

It seems that mean, se and worst portray similar correlations for each feature in the correlation matrix. The features that tend to have a higher correlation with diagnosis_M of malignant tumours are: concave points, concavity and compactness. On the other hand, it seems like the most 'irrelevant' features which do not at all indicate the status of a tumour are: symmetry, texture and fractal_dimension.

```python
import matplotlib.pyplot as plt
import random
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, KFold, GridSearchCV
from scipy.optimize import curve_fit
from sklearn.metrics import mean_squared_error, r2_score
import warnings
warnings.filterwarnings('ignore')

from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
```

```python
feature_cols = ['concave points_worst', 'concavity_worst', 'compactness_worst', 'concave points_mean', 'concavity_mean', 'compactness_mean']
# initially considered removing either the worst or mean features however removing them resulted in a lower overall preseion and accuracy score.

X = cancer_encoded[feature_cols]
y = cancer_encoded.diagnosis_M

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=3)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

rf_clf = RandomForestClassifier(random_state=1)
rf_clf.fit(X_train_scaled, y_train)

y_pred = rf_clf.predict(X_test_scaled)

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
# Import necessary libraries
from sklearn.tree import export_graphviz
from six import StringIO
from IPython.display import Image
import pydotplus

single_tree = rf_clf.estimators_[0]

dot_data = StringIO()
export_graphviz(single_tree, out_file=dot_data,
                filled=True, rounded=True,
                special_characters=True,
                feature_names=feature_cols, class_names=['0', '1'])


graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('cancer.png')
Image(graph.create_png())
```

```python
from sklearn import metrics

cm = metrics.confusion_matrix(y_test, y_pred)
cm

TN = cm.ravel()[0]
FP = cm.ravel()[1]
FN = cm.ravel()[2]
TP = cm.ravel()[3]


precision= TP/(TP+FP)
recall = TP/(TP+FN)

print('precision ', precision)
print('recall ', recall)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)

# create heatmap
sns.heatmap(pd.DataFrame(cm), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
from sklearn.metrics import classification_report
target_names = ['diagnosis_M', 'diagnosis_B']
print(classification_report(y_test, y_pred, target_names=target_names))
```
